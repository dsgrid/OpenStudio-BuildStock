```python
field_x = 'CR'
field_y1 = 'Floor Area Weighted'
field_y2 = 'HouseCountScaled'
df = pandas.DataFrame([recs.microdata.groupby(field_x)[field_y1].sum() / recs.microdata.groupby(field_x)[field_y2].sum(),
                       hector_sims_old.groupby(field_x)[field_y1].sum() / hector_sims_old.groupby(field_x)[field_y2].sum(),
                       hector_sims.groupby(field_x)[field_y1].sum() / hector_sims.groupby(field_x)[field_y2].sum()]).T
df.columns = ['RECS','ResStock (old)','ResStock']
df.plot(kind='line')
plt.gca().set_ylabel('Conditioned Floor Area')
```
